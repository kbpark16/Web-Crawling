```python
#selenium의 webdriver로 크롬 브라우저를 실행한다
driver = webdriver.Chrome('C:\\Users\\aa\\Downloads\\chromedriver_win32\\chromedriver.exe')
```

```python
#검색어를 입력한다
for i in dt['title']:
    #"Google"에 접속한다
    driver.get("https://music.bugs.co.kr/chart?wl_ref=M_left_02_01")
    #검색 입력 부분에 커서를 올리고
    #검색 입력 부분에 다양한 명령을 내리기 위해 elem 변수에 할당한다
    WebElement = driver.find_element_by_id("headerSearchInput")
    
    #print(i)
    
    WebElement.clear()
    WebElement.send_keys(i)
    
    #검색을 실행한다
    #elem.submit()
    WebElement.send_keys(Keys.RETURN)
    
    #pre_elem = driver.find_element_by_class_name("musicPDTag")
    #print(pre_elem.text)
    try:
        pre_elem = driver.find_element_by_class_name("musicPDTag")
        tag_i=pre_elem.text
        result.loc[i]=[i, tag_i]
        #result.append(pre_elem.text)
        
    except NoSuchElementException:
        result.loc[i]=[i, 'none']
        #tag.append('none')
        print('[예외발생] continue')
        continue

    #if pre_elem:
    #    tag_i=pre_elem.text
    #    result.loc[i]=[i, tag_i]
    #else:
    #    print(i)
    
    #result.append(pre_elem.text)
```
