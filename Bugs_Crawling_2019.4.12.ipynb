## 0.벅스뮤직 주간 TOP1-100, 최근 10년치 Title, aritst 크롤링
* 주의, 알고리즘 상 7일간 반복되는 주간 차트를
* 중복되게 1일씩 7번 크롤링하므로, 중복제거가 필수적임
* output 형식: txt, 날짜, 순위, 가수이름(artist), 노래제목(title)

```python
#2010년1월1일부터 현재 (2019년3월31일까지) 크롤링
#-*-coding:utf8*-
from urllib.request import urlopen
from bs4 import BeautifulSoup

f=open('bugschart_20100101_20190331.txt','w', encoding='utf-8') #파일 쓸 준비

for y in range(2010,2020): # 2010년부터 2019년까지 range라 2019+1
        firstMonth=1
        finalMonth=13 #finalMonth: 밑의 for문 range라 하한 +1 = 12월+1
        if y==2010:
                firstMonth=1 #2016년은 6월부터 처리하기 위해 설정
        elif y==2019:
                finalMonth=4 #2019년은 3월까지 처리하기 위해 설정-range하한이라+1
        for m in range(firstMonth,finalMonth):
                firstDay=1
                finalDay=32 #32+1
                if y==2010 and m==1: #y,m은 실제값, 위의 range에 들어가면 하한 +1
                        firstDay=1 #2016년은 6월 1일부터 처리하기 위해 설정
                elif y==2019 and m==3: #실제값 2019년3월일때, finalDay31일까지: 31+1
                        finalDay=32  #2019년 3월31일까지 수집

                for d in range(firstDay,finalDay):
                        if(m==2 and d>28): #2월은 28일까지로 가정
                                break
                        elif(m in [4,6,9,11] and d>30):
                                break

                        if len(str(m)) < 2: #월 자릿수가 2자리 미만 시 앞에 '0'추가
                                mStamp="0"+str(m)
                        else :
                                mStamp=str(m)
                        if len(str(d)) < 2: #일 자릿수가 2자리 미만 시 앞에 '0'추가
                                dStamp="0"+str(d)
                        else :
                                dStamp=str(d)
                                        
                        timestamp = str(y)+str(mStamp)+str(dStamp)
                        #print("날짜 : "+timestamp) , week: 월간, 같은 주면 똑같은 차트가 나옴, - 이 부분 handling 해야함
                        url=urlopen("http://music.bugs.co.kr/chart/track/week/total?chartdate="+timestamp)
                        soup=BeautifulSoup(url,"lxml", from_encoding="utf8")
                        artists=[] #이 날 1위~100위에 등록된 가수들 목록
                        artistRank=0
                        titles=[] #이 날 1위~100위에 등록된 제목 목록
                        titleRank=0
                        try:
                                for link1 in soup.find_all(name="p",attrs={"class":"artist"}):
                                        try:
                                                artist=link1.find('a').text
                                                artists.append(artist)
                                                artistRank+=1
                                                #print('날짜:'+str(timestamp)+', 순위:'+str(artistRank)+', 가수:'+str(artist))
                                        except AttributeError as artistError: #가수데이터가 존재하지 않을 경우 a태그가 없더라.
                                                artist='N/A'
                                                artists.append(artist)
                                                artistRank+=1
                                                #print('서비스 되지 않는 아티스트입니다. '+str(artistRank)+"위")
                                
                                for link2 in soup.find_all(name="p",attrs={"class":"title"}):
                                        try:
                                                title=link2.find('a').text
                                                titles.append(title)
                                                titleRank+=1
                                                #print('날짜:'+str(timestamp)+', 순위:'+str(titleRank)+', 제목:'+str(title))
                                        except AttributeError as titleError: #곡 데이터가 존재하지 않을 경우 a태그가 없더라.
                                                title='N/A'
                                                titles.append(title)
                                                titleRank+=1
                                                #print('서비스 되지 않는 곡입니다. '+str(titleRank)+"위")

                                if(artistRank!=titleRank): #혹시나 아티스트와 곡 목록 갯수가 다르면 에러를 일으킨다. (디버깅용)
                                        raise NotImplementedError

                                for i in range(0,100):
                                        f.write(str(timestamp)+','+str(i+1)+','+str(artists[i])+','+str(titles[i])+'\n')
                                        
                        except AttributeError as e: #p태그 자체가 존재하지 않을 경우, 데이터 없는것으로 여기고 다음 루프로 넘어가게 한다.
                                print(timestamp+"이 날 데이터가 존재하지 않습니다.")
                                continue
                        except NotImplementedError as notImpelemtned: #위에서 정의한 디버깅용 에러가 나면 표시해준다.
                                print('아티스트, 제목 최종랭킹 불일치')
                        except IndexError as index: #웹페이지 자체 에러로 Top100 곡 갯수가 100개가 안되면 인덱스에러로 표시한다.
                                print('인덱스 에러 / '+'아티스트 리스트 길이 : '+str(len(artists))+'/ 곡 리스트 길이 : '+str(len(titles)))
                                
f.close()
```

## 1. 각 곡에 Unique한 Track_id 뽑아오기 - 가사별로 url이 Track_id 부분만 바꿔주면 되기 때문

* 주간 차트 URL 7일 주기로 만들기

```python
head='https://music.bugs.co.kr/chart/track/week/total?chartdate='
```

* 주간 차트 URL 7일 주기를 이용하여, 2016년 6월-2019년 3월까지 TOP1-TOP100의 Track_id 긁어오기

```python
dt_index = pd.date_range(start='20100101', end='20190325' ,freq='W') # 주기 7일
# pandas.date_range(start='20160901', end='20161031',freq='W-MON')
# 을 하면 해당 기간 매주 월요일들만 추출합니다.

# type(dt_index) => DatetimeIndex
# DatetimeIndex => list(str)
dt_list = dt_index.strftime("%Y%m%d").tolist()

for i in dt_list:
    print(i)
```

```python
#track id 가져옴 - 가사 크롤링 하려고,
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import re
track_id=[]

# 2010년부터 2019년까지 모든 TOP1-100의 TRACK_ID 부터 가져오기
# https://music.bugs.co.kr/track/1902268?wl_ref=list_tr_08_chart
# track_id만 저기에 넣고 가사 크롤링 자동화 하면 됨.
for i in dt_list:
    req = requests.get(head+i)
    html = req.content
    soup = BeautifulSoup(html, 'lxml') # pip install lxml
    #print(soup.body.find('div', attrs={'class':'lyricsContainer'}))
    a=soup.find_all('td', attrs={'class':'check'})
    track_id=[track_id,a]
    regex = re.compile('[0-9]+')
    mc = regex.findall(str(track_id))
    aaa=[i for i in mc if len(i) > 5]
    print(i)
```

```python
# 위에서 얻은 track_id를 pd.dataframe으로 변형
df = pd.DataFrame({'track_id':aaa})
print (df)

# 7일 주기로 뽑았어도, 같은 곡이 여러번 나왔을 수도 있으므로,
# track_id(곡에 unique한 key)기준으루 중복제거
df_unique=df.drop_duplicates(["track_id"])

print(df_unique)
```

```python
# 위에서 얻은 unique한 track_id를, 아래의 lyrics, title을 같이 긁어올(lyrics가 목적,title은 원래
# title, artist가 붙어있는 데이터프레임과의 join key용 으로 긁어오는것
new_track_list=df_unique['track_id'].tolist()
len(new_track_list)
new_track_list[1]
```

## 3. artist + title + tag(윤선씨) 가 보내준 데이터셋 불러오기
## 여기 데이터셋에 위에서 긁어온 lyrics를 붙이려고 함

```python
#artist,title을 모아둔 원래 데이터셋 불러오기
#10761 obs, 4 columns + tag(윤선씨가 보내준 칼럼)
# 1.date, 2.arist, 3.title, 4.tag
#주의: utf-8 에러나면, 메모장으로 열어서, 인코딩 utf-8로 바꿔서 같은 이름으로 저장하고, csv로 불러오기
origin = pd.read_csv('bugs_tag_file_yunseon[tag].csv', sep=",", header=None,skiprows=1, names=["date", "artist", "title", "tag"],encoding = "utf-8-sig")
origin
```

```python
#origin 데이터에, 앞서 뽑은 lyrics_null, title_null을 merge한 'lyrics_title'데이터에
# title을 공통 키로 잡고 left - join
merged_data_lyrics=pd.merge(origin,lyrics_title,on='title')
#right join해서 index 칼럼 NaN이면 지우면 됨 ?
merged_data_lyrics=merged_data_lyrics.dropna(how='any')
merged_data_lyrics

#title 기준 unique한 것만 남기기
merged_data_lyrics_unique=merged_data_lyrics.drop_duplicates(["title","index"],keep='first')
#csv파일로 내보내기, lyrics에서 \r\n등은 csv로 내보내면 아예 없어짐
merged_data_lyrics_unique.to_csv(r'C:\Users\korea\Desktop\텍스트마이닝(강필성 교수님)\merged5.csv', index = None, header=True,encoding = "utf-8-sig")
```
